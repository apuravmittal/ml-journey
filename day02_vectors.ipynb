```python
import numpy as np
import matplotlib.pyplot as plt

# ── 1. Vector addition ───────────────────────────────────
def vec_add(a, b):
    return np.array(a) + np.array(b)

print(vec_add([1,2,3], [4,5,6]))   # → [5 7 9]

# ── 2. Scalar multiplication ─────────────────────────────
def vec_scale(a, scalar):
    return np.array(a) * scalar

print(vec_scale([1,2,3], 3))       # → [3 6 9]

# ── 3. Dot product (manual, no np.dot) ───────────────────
def dot_product(a, b):
    return sum(x*y for x,y in zip(a,b))

print(dot_product([1,2,3],[4,5,6]))  # → 32
print(np.dot([1,2,3],[4,5,6]))       # → 32  (verify they match)

# ── 4. Vector magnitude ──────────────────────────────────
def magnitude(v):
    return np.sqrt(sum(x**2 for x in v))

print(magnitude([3, 4]))             # → 5.0  (3-4-5 triangle)
print(magnitude([1, 2, 3]))          # → 3.7416...

# ── 5. Unit vector (normalise) ───────────────────────────
def unit_vector(v):
    v = np.array(v, dtype=float)
    return v / magnitude(v)

uv = unit_vector([3, 4])
print(uv)                            # → [0.6, 0.8]
print(magnitude(uv))                 # → 1.0  ✓

# ── 6. Cosine similarity ─────────────────────────────────
def cosine_similarity(a, b):
    return dot_product(a,b) / (magnitude(a) * magnitude(b))

print(cosine_similarity([1,0], [1,0]))    # → 1.0  (identical)
print(cosine_similarity([1,0], [0,1]))    # → 0.0  (perpendicular)
print(cosine_similarity([1,0], [-1,0]))   # → -1.0 (opposite)
print(cosine_similarity([1,2,3],[4,5,6])) # → ~0.974

# ── 7. Visualise 2D vectors ──────────────────────────────
fig, ax = plt.subplots(figsize=(5, 5))
origin = [0, 0]
vecs = {'a':[2,1],'b':[1,3],'a+b':[3,4]}
colors = ['#5B3FE8','#1D9E75','#D85A30']

for (name, v), c in zip(vecs.items(), colors):
    ax.annotate('', xy=v, xytext=origin,
                arrowprops=dict(arrowstyle='->', color=c, lw=2))
    ax.text(v[0]+0.1, v[1]+0.1, name, color=c, fontsize=12, fontweight='bold')

ax.set_xlim(-0.5, 4.5)
ax.set_ylim(-0.5, 4.5)
ax.axhline(0, color='gray', lw=0.5)
ax.axvline(0, color='gray', lw=0.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Vector addition: a + b = a+b', fontsize=12)
plt.tight_layout()
plt.show()
print("All done. Save as day02_vectors.ipynb and push to GitHub.")
```
